## üîç Step 1: Load the Dataset

We load the dataset into a DataFrame and preview the first few rows.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings for better visuals
pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv("../data/house_prices_records.csv")
```

```python
df.head()
```

## üìê Step 2: Dataset Dimensions and Column Overview

We check the shape of the dataset and inspect its column names and data types.

```python
print("Shape of the dataset:", df.shape)
print("\nColumn names:\n", df.columns.tolist())

print("\nData types and non-null counts:")
print(df.info())

print("\nDescriptive statistics:")
display(df.describe())
```

## ‚ùì Step 3: Missing Values

We identify columns that contain missing values and how many are missing in each.


```python
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
missing_values.sort_values(ascending=False)
```

## üìã Step 4: Duplicate Rows

Duplicate entries can introduce bias, so we count how many exist.


```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

## üì¶ Step 5: Outlier Detection

Using boxplots, we inspect common numerical features for outliers that might skew the model.


```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(x=df["SalePrice"], ax=axs[0])
sns.boxplot(x=df["GrLivArea"], ax=axs[1])
sns.boxplot(x=df["LotArea"], ax=axs[2])
axs[0].set_title("Sale Price")
axs[1].set_title("Above Ground Living Area")
axs[2].set_title("Lot Area")
plt.tight_layout()
plt.show()
```
